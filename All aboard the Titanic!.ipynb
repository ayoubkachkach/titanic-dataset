```python
import pandas as pd
```

```python
df_train = pd.read_csv('train.csv', index_col='PassengerId')
df_train.head()
```

```python
df_train.describe()
```

```python
df_train.isna().sum()
```

```python
dropped_cols = ['Name', 'Ticket', 'Cabin', 'Embarked']
df_train = df_train.drop(dropped_cols, axis=1)
df_train = df_train.fillna(df_train.median())
df_train.isna().sum()
```

```python
df_train.head()
```

```python
df_train = df_train.replace({'male': 0, 'female': 1})
```

```python
from sklearn.ensemble import RandomForestClassifier

# Shuffle dataset
df_train = df_train.sample(frac=1).reset_index(drop=True)

# Split into features and labels
X = df_train.drop(['Survived'], axis=1)
y = df_train.Survived

forest = RandomForestClassifier(n_estimators=1000, max_depth=4, random_state=42, oob_score=True)
forest.fit(X, y)
```

```python
forest.oob_score_
```

```python
df_test = pd.read_csv('test.csv', index_col='PassengerId')
df_test.head()
```

```python
df_test = df_test.drop(dropped_cols, axis=1)
# Fill missing values using median
df_test = df_test.fillna(df_test.median())
df_test.isna().sum()
# Replace Sex values with numbers
df_test = df_test.replace({'male': 0, 'female': 1})
df_test.head()
```

```python
y_pred = forest.predict(df_test)
```

```python
submission = pd.DataFrame({
        "PassengerId": df_test.index,
        "Survived": y_pred
    })

submission.to_csv('submission.csv', index=False)
```

```python
df_train = pd.read_csv('train.csv', index_col='PassengerId')
df_train.head()
```

```python
df_train.drop(['Ticket'], inplace=True, axis=1)
df_train.Embarked = df_train.Embarked.fillna('S')

df_train.head()
```

```python
TITLE_DICT = {
    "Capt": "Officer",
    "Col": "Officer",
    "Major": "Officer",
    "Jonkheer": "Honorific",
    "Don": "Honorific",
    "Dona": "Honorific",
    "Sir" : "Honorific",
    "Dr": "Officer",
    "Rev": "Honorific",
    "the Countess":"Honorific",
    "Mme": "Mrs",
    "Mlle": "Miss",
    "Ms": "Mrs",
    "Mr" : "Mr",
    "Mrs" : "Mrs",
    "Miss" : "Miss",
    "Master" : "Honorific",
    "Lady" : "Honorific"
}


def get_titles(names):
    # we extract the title from each name
    titles = names.map(lambda name:name.split(',')[1].split('.')[0].strip())
    titles = titles.map(lambda name:TITLE_DICT[name])
    return titles
```

```python
df_train['Title'] = get_titles(df_train.Name)
df_train.drop(['Name'], axis=1, inplace=True)
```

```python
def process_cabin(cabins):
    decks = cabins.fillna('U')
    decks = decks.map(lambda cabin:cabin[0])
    return decks
```

```python
df_train['Deck'] = process_cabin(df_train.Cabin)
df_train.drop(['Cabin'], axis=1, inplace=True)
df_train.head()
```

```python
df_train['FamilySize'] = df_train.Parch + df_train.SibSp + 1
```

```python
def process_age(df):
    ages = df.Age.fillna(df.groupby(['Sex', 'Title', 'Pclass']).Age.transform('mean'))
    return ages

df_train['Age'] = process_age(df_train)
```

```python
df_train.Sex = df_train.Sex.replace({'male':0, 'female':1})
df_train = pd.get_dummies(df_train)
df_train.columns
```

```python
def process_all(df):
    # Drop Ticket row
    df.drop(['Ticket'], inplace=True, axis=1)
    # Fill NA for Embarked by S
    df.Embarked = df.Embarked.fillna('S')
    # Get titles and drop Name
    df['Title'] = get_titles(df.Name)
    df.drop(['Name'], axis=1, inplace=True)
    # Extract deck number from cabin
    df['Deck'] = process_cabin(df.Cabin)
    df.drop(['Cabin'], axis=1, inplace=True)
    # Create family size
    df['FamilySize'] = df.Parch + df.SibSp + 1
    # Fill missing age values
    df['Age'] = process_age(df)
    # Fill missing fare values
    df['Fare'] = df.Fare.fillna(df.Fare.mean())
    # Replace categorical values
    df.Sex = df.Sex.replace({'male':0, 'female':1})
    df = pd.get_dummies(df)
    
    return df
```

```python
df_train = pd.read_csv('train.csv', index_col='PassengerId')

X_train = df_train.drop(['Survived'], axis=1)
y_train = df_train['Survived']
training_size = len(X_train)

X_test = pd.read_csv('test.csv', index_col='PassengerId')

df = pd.concat([X_train, X_test])
df = process_all(df)

X_train = df[:training_size]
X_test = df[training_size:]
```

```python
from sklearn.model_selection import ParameterGrid

parameter_grid = {
             'max_depth' : [2, 4, 6],
             'n_estimators': [100, 400, 600, 800, 1000],
             'max_features': ['sqrt', 'log2'],
             }

forest = RandomForestClassifier(oob_score=True)
all_parameters = ParameterGrid(parameter_grid)
best_score = 0
for parameters in all_parameters:
    forest.set_params(**parameters)
    forest.fit(X_train, y_train)
    if forest.oob_score_ > best_score:
        best_score = forest.oob_score_
        best_parameters = parameters

print("Grid: ", best_parameters)
```

```python
forest.set_params(**best_parameters)
y_pred = forest.predict(X_test)

submission = pd.DataFrame({
        "PassengerId": X_test.index,
        "Survived": y_pred
    })

submission.to_csv('submission.csv', index=False)
```
